# Report Average Output Every Second

To calculate number of values per second:

    Calculate number of values per run = total n values / 20 runs.
    
    Calculate number of values per second = n values per run / n seconds.
    
3 iterations = 90 seconds training, 4 = 120, 5 = 50, 6 = 180

For test runs, always 30 seconds

```python
goal_training_3=goal_training_3.sort_index()
goal_training_4=goal_training_4.sort_index()
goal_training_5=goal_training_5.sort_index()
goal_training_6=goal_training_6.sort_index()

goaltrain3_means = np.mean(goal_training_3.values.reshape(-1, 3), axis=1) #300/90=3.33 so ave every 3 values
goaltrain4_means = np.mean(goal_training_4.values.reshape(-1, 4), axis=1) #400/120=3.33 so ave every 4 values
goaltrain5_means = np.mean(goal_training_5.values.reshape(-1, 5), axis=1) #500/150=3.33 so ave every 5 values
goaltrain6_means = np.mean(goal_training_6.values.reshape(-1, 6), axis=1) #600/180=3.33 so ave every 6 values

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True, figsize=(15, 4))
f.suptitle('Classification of Goal Training Patterns in each Training Length Condition')
ax1.hist(np.hstack(goaltrain3_means), bins=bins, rwidth=0.95, alpha=0.5, label='90 seconds')
ax1.set_title('90 seconds')
ax2.hist(np.hstack(goaltrain4_means), bins=bins, rwidth=0.95, alpha=0.5, label='120 seconds')
ax2.set_title('120 seconds')
ax3.hist(np.hstack(goaltrain5_means), bins=bins, rwidth=0.95, alpha=0.5, label='150 seconds')
ax3.set_title('150 seconds')
ax4.hist(np.hstack(goaltrain6_means), bins=bins, rwidth=0.95, alpha=0.5, label='180 seconds')
ax4.set_title('180 seconds')
```
